### Intensity modulation (IM) with Pulse Amplitude Modulation (PAM)

```python
# simulation parameters
SpS = 16            # samples per symbol
M = 4              # order of the modulation format
Rs = 10e9          # Symbol rate (for OOK case Rs = Rb)
Fs = SpS*Rs        # Sampling frequency in samples/second
Ts = 1/Fs          # Sampling period

# Laser power
Pi_dBm = 0         # laser optical power at the input of the MZM in dBm
Pi = dBm2W(Pi_dBm) # convert from dBm to W

# MZM parameters
paramMZM = parameters()
paramMZM.Vpi = 2
paramMZM.Vb = -paramMZM.Vpi/2

# generate pseudo-random bit sequence
bitsTx = np.random.randint(2, size=int(np.log2(M)*1e6))

# generate ook modulated symbol sequence
symbTx = modulateGray(bitsTx, M, 'pam')
symbTx = pnorm(symbTx) # power normalization

# upsampling
symbolsUp = upsample(symbTx, SpS)

# typical NRZ pulse
pulse = pulseShape('nrz', SpS)
pulse = pulse/max(abs(pulse))

# pulse shaping
sigTx = firFilter(pulse, symbolsUp)

# optical modulation
Ai = np.sqrt(Pi)
sigTxo = mzm(Ai, 0.25*sigTx, paramMZM)

print('Average power of the modulated optical signal [mW]: %.3f mW'%(signal_power(sigTxo)/1e-3))
print('Average power of the modulated optical signal [dBm]: %.3f dBm'%(10*np.log10(signal_power(sigTxo)/1e-3)))

fig, axs = plt.subplots(1, 2, figsize=(16,3))
interval = np.arange(16*20,16*50)
t = interval*Ts/1e-9

# plot psd
axs[0].set_xlim(-3*Rs,3*Rs);
axs[0].set_ylim(-180,-80);
axs[0].psd(sigTx,Fs=Fs, NFFT = 16*1024, sides='twosided', label = 'RF signal spectrum')
axs[0].legend(loc='upper left');

axs[1].plot(t, sigTx[interval], label = 'RF binary signal', linewidth=2)
axs[1].set_ylabel('Amplitude (a.u.)')
axs[1].set_xlabel('Time (ns)')
axs[1].set_xlim(min(t),max(t))
axs[1].legend(loc='upper left')
axs[1].grid()

fig, axs = plt.subplots(1, 2, figsize=(16,3))
# plot psd
axs[0].set_xlim(-3*Rs,3*Rs);
axs[0].set_ylim(-255,-155);
axs[0].psd(np.abs(sigTxo)**2, Fs=Fs, NFFT = 16*1024, sides='twosided', label = 'Optical signal spectrum')
axs[0].legend(loc='upper left');

axs[1].plot(t, np.abs(sigTxo[interval])**2, label = 'Optical modulated signal', linewidth=2)
axs[1].set_ylabel('Power (p.u.)')
axs[1].set_xlabel('Time (ns)')
axs[1].set_xlim(min(t),max(t))
axs[1].legend(loc='upper left')
axs[1].grid()
```

### Linear fiber channel model (fiber + EDFA opt. amplifier)

```python
# linear optical channel
paramCh = parameters()
paramCh.L = 40         # total link distance [km]
paramCh. = 0.2        # fiber loss parameter [dB/km]
paramCh.D = 16         # fiber dispersion parameter [ps/nm/km]
paramCh.Fc = 193.1e12  # central optical frequency [Hz]
paramCh.Fs = Fs        # simulation sampling frequency [samples/second]

sigCh = linearFiberChannel(sigTxo, paramCh)

# receiver pre-amplifier
paramEDFA = parameters()
paramEDFA.G = paramCh.*paramCh.L    # edfa gain
paramEDFA.NF = 4.5   # edfa noise figure
paramEDFA.Fc = paramCh.Fc
paramEDFA.Fs = Fs

sigCh = edfa(sigCh, paramEDFA)
```

### Direct-detection (DD) pin receiver model

```python
# ideal photodiode (noiseless, no bandwidth limitation)
paramPD = parameters()
paramPD.ideal = True
paramPD.Fs = Fs
I_Rx_ideal = photodiode(sigTxo.real, paramPD)

# noisy photodiode (thermal noise + shot noise + bandwidth limitation)
paramPD = parameters()
paramPD.ideal = False
paramPD.B = Rs
paramPD.Fs = Fs

I_Rx = photodiode(sigCh, paramPD)

discard = 100

eyediagram(I_Rx_ideal[discard:-discard], I_Rx.size-2*discard, SpS, plotlabel='signal at Tx', ptype='fancy')
eyediagram(I_Rx[discard:-discard], I_Rx.size-2*discard, SpS, plotlabel='signal at Rx', ptype='fancy')
```

```python
I_Rx = I_Rx/np.std(I_Rx)

# capture samples in the middle of signaling intervals
symbRx = I_Rx[0::SpS]

# subtract DC level and normalize power
symbRx = symbRx - symbRx.mean()
symbRx = pnorm(symbRx)

# demodulate symbols to bits with minimum Euclidean distance
const = grayMapping(M,'pam') # get PAM constellation
Es = signal_power(const) # calculate the average energy per symbol of the PAM constellation

bitsRx = demodulateGray(np.sqrt(Es)*symbRx, M, 'pam')

discard = 100
err = np.logical_xor(bitsRx[discard:bitsRx.size-discard], bitsTx[discard:bitsTx.size-discard])
BER = np.mean(err)

#Pb = 0.5*erfc(Q/np.sqrt(2)) # theoretical error probability
print('Number of counted errors = %d '%(err.sum()))
print('BER = %.2e '%(BER))
#print('Pb = %.2e '%(Pb))

err = err*1.0
err[err==0] = np.nan

plt.plot(err,'o', label = 'bit errors')
plt.vlines(np.where(err>0), 0, 1)
plt.xlabel('bit position')
plt.ylabel('counted error')
plt.legend()
plt.grid()
plt.ylim(0, 1.5)
plt.xlim(0,err.size);
```

### Generate curve of BER vs received input power

```python
# simulation parameters
SpS = 16            # Samples per symbol
M = 4               # order of the modulation format
Rs = 40e9           # Symbol rate (for the OOK case, Rs = Rb)
Fs = SpS*Rs         # Signal sampling frequency (samples/second)
Ts = 1/Fs           # Sampling period

# MZM parameters
paramMZM = parameters()
paramMZM.Vpi = 2
paramMZM.Vb = -paramMZM.Vpi/2

# typical NRZ pulse
pulse = pulseShape('nrz', SpS)
pulse = pulse/max(abs(pulse))

# photodiode parameters
paramPD = parameters()
paramPD.ideal = False
paramPD.B = Rs
paramPD.Fs = Fs

powerValues = np.arange(-20,-4) # power values at the input of the pin receiver
BER = np.zeros(powerValues.shape)
Pb = np.zeros(powerValues.shape)

const = grayMapping(M,'pam') # get PAM constellation
Es = signal_power(const) # calculate the average energy per symbol of the PAM constellation

discard = 100
for indPi, Pi_dBm in enumerate(tqdm(powerValues)):

    Pi = dBm2W(Pi_dBm+3) # optical signal power in W at the MZM input

    # generate pseudo-random bit sequence
    bitsTx = np.random.randint(2, size=int(np.log2(M)*1e6))
    n = np.arange(0, bitsTx.size)

    # generate ook modulated symbol sequence
    symbTx = modulateGray(bitsTx, M, 'pam')
    symbTx = pnorm(symbTx) # power normalization

    # upsampling
    symbolsUp = upsample(symbTx, SpS)

    # pulse formatting
    sigTx = firFilter(pulse, symbolsUp)

    # optical modulation
    Ai = np.sqrt(Pi)
    sigTxo = mzm(Ai, 0.25*sigTx, paramMZM)

    # pin receiver
    I_Rx = photodiode(sigTxo.real, paramPD)
    I_Rx = I_Rx/np.std(I_Rx)

    # capture samples in the middle of signaling intervals
    symbRx = I_Rx[0::SpS]

    # subtract DC level and normalize power
    symbRx = symbRx - symbRx.mean()
    symbRx = pnorm(symbRx)

    snr = signal_power(symbRx)/(2*signal_power(symbRx-symbTx))
    EbN0 = 10*np.log10(snr/np.log2(M))

    # demodulate symbols to bits with minimum Euclidean distance
    bitsRx = demodulateGray(np.sqrt(Es)*symbRx, M, 'pam')

    err = np.logical_xor(bitsRx[discard:bitsRx.size-discard], bitsTx[discard:bitsTx.size-discard])
    BER[indPi] = np.mean(err)
    Pb[indPi] = theoryBER(M, EbN0, 'pam') # probability of bit error (theory)
```

```python
plt.figure()
plt.plot(powerValues, np.log10(Pb),'--',label='Pb (theory)')
plt.plot(powerValues, np.log10(BER),'o',label='BER')
plt.grid()
plt.ylabel('log10(BER)')
plt.xlabel('Pin (dBm)');
plt.title('BER vs input power at the pin receiver')
plt.legend();
plt.ylim(-10,0);
plt.xlim(min(powerValues), max(powerValues));
```

### Generate curve of BER vs transmission distance

```python
# simulation parameters
SpS = 16            # Samples per symbol
M = 4               # order of the modulation format
Rs = 40e9           # Symbol rate (for the OOK case, Rs = Rb)
Fs = SpS*Rs         # Signal sampling frequency (samples/second)
Ts = 1/Fs           # Sampling period

# Laser power
Pi_dBm = 0         # laser optical power at the input of the MZM in dBm
Pi = dBm2W(Pi_dBm) # convert from dBm to W

# MZM parameters
paramMZM = parameters()
paramMZM.Vpi = 2
paramMZM.Vb = -paramMZM.Vpi/2

# typical NRZ pulse
pulse = pulseShape('nrz', SpS)
pulse = pulse/max(abs(pulse))

# fiber channel parameters
distance = np.arange(0,12.5,0.5) # transmission distance in km
paramCh = parameters()
paramCh. = 0.2        # fiber loss parameter [dB/km]
paramCh.D = 16         # fiber dispersion parameter [ps/nm/km]
paramCh.Fc = 193.1e12  # central optical frequency [Hz]
paramCh.Fs = Fs        # simulation sampling frequency [samples/second]

# receiver pre-amplifier parameters
paramEDFA = parameters()
paramEDFA.NF = 4.5   # edfa noise figure
paramEDFA.Fc = paramCh.Fc
paramEDFA.Fs = Fs

sigCh = edfa(sigCh, paramEDFA)

# photodiode parameters
paramPD = parameters()
paramPD.ideal = False
paramPD.B = Rs
paramPD.Fs = Fs

BER = np.zeros(distance.shape)
Pb = np.zeros(distance.shape)

const = grayMapping(M,'pam') # get PAM constellation
Es = signal_power(const) # calculate the average energy per symbol of the PAM constellation

discard = 100
for indL, L in enumerate(tqdm(distance)):

    # generate pseudo-random bit sequence
    bitsTx = np.random.randint(2, size=int(np.log2(M)*1e5))
    n = np.arange(0, bitsTx.size)

    # generate ook modulated symbol sequence
    symbTx = modulateGray(bitsTx, M, 'pam')
    symbTx = pnorm(symbTx) # power normalization

    # upsampling
    symbolsUp = upsample(symbTx, SpS)

    # pulse formatting
    sigTx = firFilter(pulse, symbolsUp)

    # optical modulation
    Ai = np.sqrt(Pi)*np.ones(sigTx.size)
    sigTxo = mzm(Ai, 0.25*sigTx, paramMZM)

    # linear optical channel
    paramCh.L = L
    sigCh = linearFiberChannel(sigTxo, paramCh)

    # receiver pre-amplifier
    if L > 0:
        paramEDFA.G = paramCh.*L  # edfa gain
        sigCh = edfa(sigCh, paramEDFA)

    # pin receiver
    I_Rx = photodiode(sigCh, paramPD)
    I_Rx = I_Rx/np.std(I_Rx)

    # capture samples in the middle of signaling intervals
    symbRx = I_Rx[0::SpS]

    # subtract DC level and normalize power
    symbRx = symbRx - symbRx.mean()
    symbRx = pnorm(symbRx)

    snr = signal_power(symbRx)/(2*signal_power(symbRx-symbTx))
    EbN0 = 10*np.log10(snr/np.log2(M))

    # demodulate symbols to bits with minimum Euclidean distance
    bitsRx = demodulateGray(np.sqrt(Es)*symbRx, M, 'pam')

    err = np.logical_xor(bitsRx[discard:bitsRx.size-discard], bitsTx[discard:bitsTx.size-discard])
    BER[indL] = np.mean(err)
    Pb[indL] = theoryBER(M, EbN0, 'pam') # probability of bit error (theory)
```

```python
plt.figure()
plt.plot(distance, np.log10(Pb),'--',label='Pb (theory)')
plt.plot(distance, np.log10(BER),'o',label='BER')
plt.grid()
plt.ylabel('log10(BER)')
plt.xlabel('Distance (km)');
plt.title('BER vs transmission distance')
plt.legend();
plt.ylim(-10,0);
plt.xlim(min(distance), max(distance));
```

```python
# Create arrays to store results for different dispersion values and modulation formats
dispersion_values = np.array([0, 5, 10, 16, 20, 25, 30])  # ps/nm/km
modulation_formats = [4, 8]  # PAM-4 and PAM-8 only
max_distances = np.zeros((len(dispersion_values), len(modulation_formats)))
all_BER_results = {}  # Dictionary to store all BER vs distance results

# Outer loops for dispersion values and modulation formats
for d_idx, D_value in enumerate(tqdm(dispersion_values)):
    for m_idx, M_value in enumerate(modulation_formats):

        # Update dispersion parameter
        paramCh.D = D_value
        M = M_value

        # Initialize BER array for this dispersion and modulation combination
        BER = np.zeros(distance.shape)

        # Run the existing distance loop (your original code remains here)
        for indL, L in enumerate(tqdm(distance)):
            # generate pseudo-random bit sequence
            bitsTx = np.random.randint(2, size=int(np.log2(M)*1e5))
            n = np.arange(0, bitsTx.size)

            # generate ook modulated symbol sequence
            symbTx = modulateGray(bitsTx, M, 'pam')
            symbTx = pnorm(symbTx) # power normalization

            # upsampling
            symbolsUp = upsample(symbTx, SpS)

            # pulse formatting
            sigTx = firFilter(pulse, symbolsUp)

            # optical modulation
            Ai = np.sqrt(Pi)*np.ones(sigTx.size)
            sigTxo = mzm(Ai, 0.25*sigTx, paramMZM)

            # linear optical channel
            paramCh.L = L
            sigCh = linearFiberChannel(sigTxo, paramCh)

            # receiver pre-amplifier
            if L > 0:
                paramEDFA.G = paramCh.*L  # edfa gain
                sigCh = edfa(sigCh, paramEDFA)

            # pin receiver
            I_Rx = photodiode(sigCh, paramPD)
            I_Rx = I_Rx/np.std(I_Rx)

            # capture samples in the middle of signaling intervals
            symbRx = I_Rx[0::SpS]

            # subtract DC level and normalize power
            symbRx = symbRx - symbRx.mean()
            symbRx = pnorm(symbRx)

            snr = signal_power(symbRx)/(2*signal_power(symbRx-symbTx))
            EbN0 = 10*np.log10(snr/np.log2(M))

            # demodulate symbols to bits with minimum Euclidean distance
            const = grayMapping(M,'pam') # get PAM constellation
            Es = signal_power(const) # calculate the average energy per symbol
            bitsRx = demodulateGray(np.sqrt(Es)*symbRx, M, 'pam')

            err = np.logical_xor(bitsRx[discard:bitsRx.size-discard], bitsTx[discard:bitsTx.size-discard])
            BER[indL] = np.mean(err)

        # After completing the distance sweep, find max distance where BER < 10^-3
        below_threshold = np.where(BER < 1e-3)[0]
        if len(below_threshold) > 0:
            max_distances[d_idx, m_idx] = distance[below_threshold[-1]]
        else:
            max_distances[d_idx, m_idx] = 0

        # Store BER results for plotting
        all_BER_results[f'D{D_value}_M{M_value}'] = BER.copy()

# Print numerical values for tables
print("\n----- Maximum Achievable Distance (km) at BER < 10^-3 -----")
print("Dispersion (ps/nm/km) | PAM-4 | PAM-8")
print("-" * 45)
for d_idx, D_value in enumerate(dispersion_values):
    print(f"{D_value:18.1f} | {max_distances[d_idx, 0]:5.1f} | {max_distances[d_idx, 1]:5.1f}")

# Print BER values at specific distances for each configuration
print("\n----- BER Values at Different Distances -----")
distances_to_report = [0, 2, 4, 6, 8, 10]  # Select specific distances to report
for d_value in dispersion_values:
    print(f"\nDispersion = {d_value} ps/nm/km")
    print("Distance (km) | PAM-4 BER | PAM-8 BER")
    print("-" * 45)
    for dist in distances_to_report:
        if dist in distance:
            dist_idx = np.where(distance == dist)[0][0]
            ber_pam4 = all_BER_results.get(f'D{d_value}_M4', np.zeros(len(distance)))[dist_idx]
            ber_pam8 = all_BER_results.get(f'D{d_value}_M8', np.zeros(len(distance)))[dist_idx]
            print(f"{dist:13.1f} | {ber_pam4:.2e} | {ber_pam8:.2e}")

# Plot maximum distance vs modulation format for each dispersion value
plt.figure(figsize=(10, 6))
for d_idx, D_value in enumerate(dispersion_values):
    plt.plot(modulation_formats, max_distances[d_idx, :], 'o-',
             label=f'D = {D_value} ps/nm/km')
plt.xlabel('Modulation Format (PAM-M)')
plt.ylabel('Maximum Achievable Distance (km)')
plt.title('Maximum Distance vs Modulation Format for Different Dispersion Values (BER < 10^-3)')
plt.legend()
plt.grid(True)
plt.show()

# Plot BER vs distance for each modulation format
for m_idx, M_value in enumerate(modulation_formats):
    plt.figure(figsize=(10, 6))
    for d_idx, D_value in enumerate(dispersion_values):
        key = f'D{D_value}_M{M_value}'
        if key in all_BER_results:
            plt.semilogy(distance, all_BER_results[key],
                       label=f'D = {D_value} ps/nm/km')
    plt.xlabel('Distance (km)')
    plt.ylabel('BER')
    plt.title(f'BER vs Distance for PAM-{M_value} with Different Dispersion Values')
    plt.axhline(y=1e-3, color='r', linestyle='--', label='BER = 10^-3')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot maximum distance vs dispersion for different modulation formats
plt.figure(figsize=(10, 6))
for m_idx, M_value in enumerate(modulation_formats):
    plt.plot(dispersion_values, max_distances[:, m_idx], 'o-',
             label=f'PAM-{M_value}')
plt.xlabel('Dispersion (ps/nm/km)')
plt.ylabel('Maximum Achievable Distance (km)')
plt.title('Maximum Achievable Distance vs Dispersion for Different PAM Formats (BER < 10^-3)')
plt.legend()
plt.grid(True)
plt.show()

# Export data to CSV for further analysis (optional)
np.savetxt('max_distances.csv',
           np.column_stack((dispersion_values, max_distances)),
           delimiter=',',
           header='Dispersion,PAM-4,PAM-8',
           comments='')
```
