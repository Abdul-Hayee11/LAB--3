```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

class parameters:
    def __init__(self):
        self.SpS_in = 4   # Samples per symbol input
        self.SpS_out = 16  # Samples per symbol output
        self.Rs = 100     # Symbol rate

# Rational resampling function
def resample(signal, param):
    """
    Perform rational resampling on a signal

    Parameters:
    ----------
    signal : numpy array
        Input signal to be resampled
    param : parameters object
        Contains SpS_in, SpS_out, and Rs parameters

    Returns:
    -------
    numpy array
        Resampled signal
    """
    # Determine the resampling factor
    resFactor = param.SpS_out / param.SpS_in

    # Create the new time vector
    n_samples_new = int(len(signal) * resFactor)

    # Perform the resampling (simple linear interpolation)
    indices = np.linspace(0, len(signal) - 1, n_samples_new)
    resampled_signal = np.interp(indices, np.arange(len(signal)), signal)

    return resampled_signal

# Clock sampling interpolation
def clockSamplingInterp(signal, Fs_in, Fs_out, jitter_rms=0):
    """
    Perform clock sampling interpolation on a signal

    Parameters:
    ----------
    signal : numpy array
        Input signal to be resampled
    Fs_in : float
        Input sampling frequency
    Fs_out : float
        Output sampling frequency
    jitter_rms : float
        RMS jitter to add to sampling times

    Returns:
    -------
    numpy array
        Resampled signal with adjusted clock
    """
    t_in = np.arange(len(signal)) / Fs_in

    # Calculate the number of samples in the output signal
    t_duration = t_in[-1]
    n_samples_out = int(np.ceil(t_duration * Fs_out))

    # Create ideal sampling times
    t_out_ideal = np.arange(n_samples_out) / Fs_out

    # Add jitter if requested
    if jitter_rms > 0:
        jitter = np.random.normal(0, jitter_rms, n_samples_out)
        t_out = t_out_ideal + jitter
    else:
        t_out = t_out_ideal

    # Only keep samples within input time range
    t_out = t_out[t_out <= t_in[-1]]

    # Interpolate signal values at new sampling times
    resampled_signal = np.interp(t_out, t_in, signal.flatten())

    return t_out, resampled_signal

# Quantizer function
def quantizer(signal, nBits):
    """
    Quantize a signal to a specified number of bits

    Parameters:
    ----------
    signal : numpy array
        Input signal to be quantized
    nBits : int
        Number of bits for quantization

    Returns:
    -------
    numpy array
        Quantized signal
    """
    # Determine the number of quantization levels
    nLevels = 2**nBits

    # Normalize signal to range [-1, 1]
    signal_max = np.max(np.abs(signal))
    signal_norm = signal / signal_max if signal_max > 0 else signal

    # Quantize the normalized signal
    step_size = 2.0 / nLevels
    quantized = np.round(signal_norm / step_size) * step_size

    # Clip to ensure values are within [-1, 1]
    quantized = np.clip(quantized, -1.0 + step_size/2, 1.0 - step_size/2)

    # Restore original scale
    quantized = quantized * signal_max

    return quantized

# Signal delay function
def delaySignal(signal, delay, Fs):
    """
    Apply a delay to a signal

    Parameters:
    ----------
    signal : numpy array
        Input signal to be delayed
    delay : float
        Delay in seconds
    Fs : float
        Sampling frequency in Hz

    Returns:
    -------
    numpy array
        Delayed signal
    """
    # Convert delay from seconds to samples
    delay_samples = int(delay * Fs)

    # Create delayed signal
    if np.iscomplex(signal).any():
        # Handle complex signals
        delayed_signal = np.zeros_like(signal)
        delayed_signal[delay_samples:] = signal[:-delay_samples] if delay_samples > 0 else signal
    else:
        # Handle real signals
        delayed_signal = np.zeros_like(signal)
        delayed_signal[delay_samples:] = signal[:-delay_samples] if delay_samples > 0 else signal

    return delayed_signal

# Function to run and plot the resampling experiment
def experiment_resampling():
    """
    Experiment with rational resampling at different rates
    """
    # Set up the figure with 4 subplots
    plt.figure(figsize=(15, 12))
    gs = GridSpec(4, 1)

    # Generate original signal
    Fs = 400  # Sampling frequency: 400 Hz
    fc = 100  # Carrier frequency: 100 Hz
    t = np.arange(0, 4096) * (1/Fs)  # Time vector
    sig = np.sin(2 * np.pi * fc * t)  # Sinusoidal signal

    # Calculate samples per cycle in original signal
    samples_per_cycle_orig = Fs / fc

    # Plot original signal
    ax1 = plt.subplot(gs[0])
    ax1.plot(t, sig, '-o', markersize=4)
    ax1.set_xlim(0, 5 * 1/fc)  # Show 5 cycles for better visibility
    ax1.set_title(f'Original Signal (Fs = {Fs} Hz, {samples_per_cycle_orig:.1f} samples/cycle)')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Amplitude')
    ax1.grid(True)

    # Resampling experiment 1: Upsampling (4 to 16 SpS) - 4x increase
    paramDec1 = parameters()
    paramDec1.SpS_in = 4
    paramDec1.SpS_out = 16
    paramDec1.Rs = fc
    resFactor1 = paramDec1.SpS_out / paramDec1.SpS_in
    Fs_new1 = Fs * resFactor1
    samples_per_cycle_1 = Fs_new1 / fc

    sig_dec1 = resample(sig, paramDec1)
    t_dec1 = np.arange(0, len(sig_dec1)) * (1/Fs_new1)

    ax2 = plt.subplot(gs[1])
    ax2.plot(t[:500], sig[:500], '-o', markersize=3, color='blue', alpha=0.5, label='Original')
    ax2.plot(t_dec1[:500*int(resFactor1)], sig_dec1[:500*int(resFactor1)], '-o', markersize=3, color='green', label='Upsampled 4x')
    ax2.set_xlim(0, 5 * 1/fc)
    ax2.set_title(f'Upsampling: {Fs} Hz  {Fs_new1} Hz ({samples_per_cycle_1:.1f} samples/cycle)')
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Amplitude')
    ax2.legend()
    ax2.grid(True)

    # Resampling experiment 2: Moderate downsampling (4 to 2 SpS) - 0.5x decrease
    paramDec2 = parameters()
    paramDec2.SpS_in = 4
    paramDec2.SpS_out = 2
    paramDec2.Rs = fc
    resFactor2 = paramDec2.SpS_out / paramDec2.SpS_in
    Fs_new2 = Fs * resFactor2
    samples_per_cycle_2 = Fs_new2 / fc

    sig_dec2 = resample(sig, paramDec2)
    t_dec2 = np.arange(0, len(sig_dec2)) * (1/Fs_new2)

    ax3 = plt.subplot(gs[2])
    ax3.plot(t[:500], sig[:500], '-o', markersize=3, color='blue', alpha=0.5, label='Original')
    ax3.plot(t_dec2[:int(500*resFactor2)], sig_dec2[:int(500*resFactor2)], '-o', markersize=4, color='orange', label='Downsampled 0.5x')
    ax3.set_xlim(0, 5 * 1/fc)
    ax3.set_title(f'Moderate Downsampling: {Fs} Hz  {Fs_new2} Hz ({samples_per_cycle_2:.1f} samples/cycle)')
    ax3.set_xlabel('Time (s)')
    ax3.set_ylabel('Amplitude')
    ax3.legend()
    ax3.grid(True)

    # Resampling experiment 3: Severe downsampling (4 to 1 SpS) - 0.25x decrease
    paramDec3 = parameters()
    paramDec3.SpS_in = 4
    paramDec3.SpS_out = 1
    paramDec3.Rs = fc
    resFactor3 = paramDec3.SpS_out / paramDec3.SpS_in
    Fs_new3 = Fs * resFactor3
    samples_per_cycle_3 = Fs_new3 / fc

    sig_dec3 = resample(sig, paramDec3)
    t_dec3 = np.arange(0, len(sig_dec3)) * (1/Fs_new3)

    ax4 = plt.subplot(gs[3])
    ax4.plot(t[:500], sig[:500], '-o', markersize=3, color='blue', alpha=0.5, label='Original')
    ax4.plot(t_dec3[:int(500*resFactor3)], sig_dec3[:int(500*resFactor3)], '-s', markersize=6, color='red', label='Downsampled 0.25x')
    ax4.set_xlim(0, 5 * 1/fc)
    ax4.set_title(f'Severe Downsampling: {Fs} Hz  {Fs_new3} Hz ({samples_per_cycle_3:.1f} samples/cycle)')
    ax4.set_xlabel('Time (s)')
    ax4.set_ylabel('Amplitude')
    ax4.legend()
    ax4.grid(True)

    plt.tight_layout()
    plt.savefig('resampling_experiment.png', dpi=300)
    plt.show()

    # Calculate and display resampling metrics
    print("\n=== Resampling Experiment Results ===")
    print(f"Original signal: Fs = {Fs} Hz, fc = {fc} Hz")
    print(f"Samples per cycle in original: {samples_per_cycle_orig:.1f}")
    print("\nUpsampling (4x):")
    print(f"  New sampling rate: {Fs_new1:.1f} Hz")
    print(f"  Samples per cycle: {samples_per_cycle_1:.1f}")
    print(f"  Impact: Increased time resolution, smoother signal reconstruction")

    print("\nModerate Downsampling (0.5x):")
    print(f"  New sampling rate: {Fs_new2:.1f} Hz")
    print(f"  Samples per cycle: {samples_per_cycle_2:.1f}")
    print(f"  Impact: Reduced data rate, still adequate representation")

    print("\nSevere Downsampling (0.25x):")
    print(f"  New sampling rate: {Fs_new3:.1f} Hz")
    print(f"  Samples per cycle: {samples_per_cycle_3:.1f}")
    print(f"  Impact: Minimal representation, approaching Nyquist limit (2 samples/cycle)")

    # Calculate reconstructed signal quality (correlation with original)
    # For the severely downsampled case, upsample it back to original rate for comparison
    paramUp = parameters()
    paramUp.SpS_in = 1
    paramUp.SpS_out = 4
    paramUp.Rs = fc
    sig_reconstructed = resample(sig_dec3, paramUp)

    # Trim to minimum common length
    min_len = min(len(sig), len(sig_reconstructed))
    correlation = np.corrcoef(sig[:min_len], sig_reconstructed[:min_len])[0,1]
    print(f"\nCorrelation between original and reconstructed from severe downsampling: {correlation:.4f}")

    # Frequency domain analysis information without the plot
    print("\nFrequency Domain Analysis:")
    print(f"Original signal primary component: {fc} Hz")
    print(f"Downsampled signal primary component still visible at {fc} Hz")
    print(f"Maximum frequency representable in downsampled signal (Nyquist): {Fs_new3/2} Hz")

# New function to demonstrate resampling on sinusoidal signals with different frequencies
def simulate_resampling_impact():
    """
    Simulate a sinusoidal signal and perform rational resampling with different sampling rates.
    Plot the original and resampled signals for comparison and analyze the impact of resampling.
    """
    plt.figure(figsize=(15, 18))
    gs = GridSpec(5, 1)

    # Parameters for the simulation
    duration = 0.1  # seconds
    Fs_original = 1000  # Original sampling frequency: 1000 Hz

    # Generate time vector for the original signal
    t_original = np.arange(0, duration, 1/Fs_original)

    # Generate sinusoidal signals with different frequencies
    f1 = 50   # Low frequency: 50 Hz
    f2 = 200  # Medium frequency: 200 Hz
    f3 = 400  # High frequency: 400 Hz (close to Nyquist)

    # Create original signals
    sig1 = np.sin(2 * np.pi * f1 * t_original)
    sig2 = np.sin(2 * np.pi * f2 * t_original)
    sig3 = np.sin(2 * np.pi * f3 * t_original)

    # Plot original signals
    ax1 = plt.subplot(gs[0])
    ax1.plot(t_original, sig1, 'b-', label=f'{f1} Hz')
    ax1.plot(t_original, sig2, 'g-', label=f'{f2} Hz')
    ax1.plot(t_original, sig3, 'r-', label=f'{f3} Hz')
    ax1.set_xlim(0, 0.05)  # Show 0.05 seconds for better visibility
    ax1.set_title(f'Original Signals at {Fs_original} Hz Sampling Rate')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Amplitude')
    ax1.legend()
    ax1.grid(True)

    # Create parameter objects for different resampling scenarios
    # 1. Upsampling to 2x the original rate
    param_up = parameters()
    param_up.SpS_in = 1
    param_up.SpS_out = 2
    Fs_up = Fs_original * (param_up.SpS_out / param_up.SpS_in)

    # 2. Downsampling to 1/2 the original rate
    param_down_moderate = parameters()
    param_down_moderate.SpS_in = 2
    param_down_moderate.SpS_out = 1
    Fs_down_moderate = Fs_original * (param_down_moderate.SpS_out / param_down_moderate.SpS_in)

    # 3. Severe downsampling to 1/5 the original rate
    param_down_severe = parameters()
    param_down_severe.SpS_in = 5
    param_down_severe.SpS_out = 1
    Fs_down_severe = Fs_original * (param_down_severe.SpS_out / param_down_severe.SpS_in)

    # Perform resampling for each signal at different rates
    # Upsampling (2x)
    sig1_up = resample(sig1, param_up)
    sig2_up = resample(sig2, param_up)
    sig3_up = resample(sig3, param_up)
    t_up = np.arange(0, len(sig1_up)) * (1/Fs_up)

    # Moderate downsampling (1/2)
    sig1_down_mod = resample(sig1, param_down_moderate)
    sig2_down_mod = resample(sig2, param_down_moderate)
    sig3_down_mod = resample(sig3, param_down_moderate)
    t_down_mod = np.arange(0, len(sig1_down_mod)) * (1/Fs_down_moderate)

    # Severe downsampling (1/5)
    sig1_down_sev = resample(sig1, param_down_severe)
    sig2_down_sev = resample(sig2, param_down_severe)
    sig3_down_sev = resample(sig3, param_down_severe)
    t_down_sev = np.arange(0, len(sig1_down_sev)) * (1/Fs_down_severe)

    # Plot upsampled signals
    ax2 = plt.subplot(gs[1])
    ax2.plot(t_up, sig1_up, 'b-', label=f'{f1} Hz')
    ax2.plot(t_up, sig2_up, 'g-', label=f'{f2} Hz')
    ax2.plot(t_up, sig3_up, 'r-', label=f'{f3} Hz')
    ax2.set_xlim(0, 0.05)
    ax2.set_title(f'Upsampled Signals (2x): {Fs_original} Hz  {Fs_up} Hz')
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Amplitude')
    ax2.legend()
    ax2.grid(True)

    # Plot moderately downsampled signals
    ax3 = plt.subplot(gs[2])
    ax3.plot(t_down_mod, sig1_down_mod, 'b-o', markersize=4, label=f'{f1} Hz')
    ax3.plot(t_down_mod, sig2_down_mod, 'g-o', markersize=4, label=f'{f2} Hz')
    ax3.plot(t_down_mod, sig3_down_mod, 'r-o', markersize=4, label=f'{f3} Hz')
    ax3.set_xlim(0, 0.05)
    ax3.set_title(f'Moderately Downsampled Signals (1/2): {Fs_original} Hz  {Fs_down_moderate} Hz')
    ax3.set_xlabel('Time (s)')
    ax3.set_ylabel('Amplitude')
    ax3.legend()
    ax3.grid(True)

    # Plot severely downsampled signals
    ax4 = plt.subplot(gs[3])
    ax4.plot(t_down_sev, sig1_down_sev, 'b-s', markersize=6, label=f'{f1} Hz')
    ax4.plot(t_down_sev, sig2_down_sev, 'g-s', markersize=6, label=f'{f2} Hz')
    ax4.plot(t_down_sev, sig3_down_sev, 'r-s', markersize=6, label=f'{f3} Hz')
    ax4.set_xlim(0, 0.05)
    ax4.set_title(f'Severely Downsampled Signals (1/5): {Fs_original} Hz  {Fs_down_severe} Hz')
    ax4.set_xlabel('Time (s)')
    ax4.set_ylabel('Amplitude')
    ax4.legend()
    ax4.grid(True)

    # Zoom in on a portion of the severely downsampled signals to see aliasing effects
    ax5 = plt.subplot(gs[4])
    ax5.plot(t_original, sig3, 'r-', alpha=0.3, label=f'Original {f3} Hz')
    ax5.plot(t_down_sev, sig3_down_sev, 'r-s', markersize=6, label=f'Downsampled {f3} Hz')

    # Add reconstructed signal from downsampled points (using interpolation)
    t_recon = t_original
    sig3_recon = np.interp(t_recon, t_down_sev, sig3_down_sev)
    ax5.plot(t_recon, sig3_recon, 'k--', linewidth=1, label='Reconstructed Signal')

    ax5.set_xlim(0, 0.03)
    ax5.set_title(f'Detailed view: {f3} Hz Signal with Severe Downsampling')
    ax5.set_xlabel('Time (s)')
    ax5.set_ylabel('Amplitude')
    ax5.legend()
    ax5.grid(True)

    plt.tight_layout()
    plt.savefig('sinusoidal_resampling_impact.png', dpi=300)
    plt.show()

    # Calculate and display sampling metrics for each frequency and sampling rate
    results = []

    def calc_samples_per_cycle(freq, fs):
        return fs / freq

    def check_nyquist_condition(freq, fs):
        return "Satisfied" if fs > 2*freq else "Violated!"

    # Original signal metrics
    results.append(f"\n=== Resampling Impact Analysis ===")
    results.append(f"Original sampling rate: {Fs_original} Hz")
    results.append(f"\nSignal Frequency: {f1} Hz")
    results.append(f"  Samples per cycle (original): {calc_samples_per_cycle(f1, Fs_original):.1f}")
    results.append(f"  Samples per cycle (upsampled): {calc_samples_per_cycle(f1, Fs_up):.1f}")
    results.append(f"  Samples per cycle (moderate down): {calc_samples_per_cycle(f1, Fs_down_moderate):.1f}")
    results.append(f"  Samples per cycle (severe down): {calc_samples_per_cycle(f1, Fs_down_severe):.1f}")
    results.append(f"  Nyquist condition at severe downsampling: {check_nyquist_condition(f1, Fs_down_severe)}")

    results.append(f"\nSignal Frequency: {f2} Hz")
    results.append(f"  Samples per cycle (original): {calc_samples_per_cycle(f2, Fs_original):.1f}")
    results.append(f"  Samples per cycle (upsampled): {calc_samples_per_cycle(f2, Fs_up):.1f}")
    results.append(f"  Samples per cycle (moderate down): {calc_samples_per_cycle(f2, Fs_down_moderate):.1f}")
    results.append(f"  Samples per cycle (severe down): {calc_samples_per_cycle(f2, Fs_down_severe):.1f}")
    results.append(f"  Nyquist condition at severe downsampling: {check_nyquist_condition(f2, Fs_down_severe)}")

    results.append(f"\nSignal Frequency: {f3} Hz")
    results.append(f"  Samples per cycle (original): {calc_samples_per_cycle(f3, Fs_original):.1f}")
    results.append(f"  Samples per cycle (upsampled): {calc_samples_per_cycle(f3, Fs_up):.1f}")
    results.append(f"  Samples per cycle (moderate down): {calc_samples_per_cycle(f3, Fs_down_moderate):.1f}")
    results.append(f"  Samples per cycle (severe down): {calc_samples_per_cycle(f3, Fs_down_severe):.1f}")
    results.append(f"  Nyquist condition at severe downsampling: {check_nyquist_condition(f3, Fs_down_severe)}")

    # Discussion of resampling impact
    results.append("\n=== Discussion of Resampling Impact ===")
    results.append("1. Upsampling (2x the original rate):")
    results.append("   - Increased time resolution and smoother signal representation")
    results.append("   - No additional frequency information is gained, but the signal appears smoother")
    results.append("   - Useful for visual analysis and when higher precision is needed in time domain")

    results.append("\n2. Moderate Downsampling (1/2 the original rate):")
    results.append("   - Reduced data rate while maintaining reasonable signal fidelity")
    results.append(f"   - Lower frequency signals ({f1} Hz) are still well-represented")
    results.append(f"   - Higher frequency signals ({f3} Hz) have fewer samples per cycle but still satisfy Nyquist")
    results.append("   - Good compromise between data rate and signal fidelity")

    results.append("\n3. Severe Downsampling (1/5 the original rate):")
    results.append("   - Significant reduction in data rate but at a cost of signal fidelity")
    results.append(f"   - Low frequency signal ({f1} Hz) remains recognizable with {calc_samples_per_cycle(f1, Fs_down_severe):.1f} samples per cycle")
    results.append(f"   - Medium frequency signal ({f2} Hz) begins to degrade with only {calc_samples_per_cycle(f2, Fs_down_severe):.1f} samples per cycle")
    results.append(f"   - High frequency signal ({f3} Hz) shows significant distortion and aliasing")
    results.append(f"   - The {f3} Hz signal has only {calc_samples_per_cycle(f3, Fs_down_severe):.1f} samples per cycle, which violates Nyquist criterion")

    results.append("\n4. Aliasing Effects:")
    results.append(f"   - The {f3} Hz signal with severe downsampling demonstrates aliasing")
    results.append(f"   - Original frequency ({f3} Hz) appears at a lower frequency in the downsampled signal")
    results.append(f"   - This is because sampling rate ({Fs_down_severe} Hz) is below the Nyquist frequency ({f3*2} Hz)")
    results.append("   - Reconstructed signal from severely downsampled points misrepresents the original")

    results.append("\n5. Practical Implications:")
    results.append("   - Signal bandwidth must be considered when choosing sampling rate")
    results.append("   - Upsampling is useful for interpolation and visual presentation")
    results.append("   - Moderate downsampling can be used to reduce data rate while preserving signal fidelity")
    results.append("   - Anti-aliasing filters should be applied before severe downsampling to prevent aliasing")

    # Print the analysis
    for line in results:
        print(line)

# Function to run all experiments
def run_all_experiments():
    """
    Run all DSP experiments
    """
    print("="*80)
    print("Running Original Resampling Experiment...")
    print("="*80)
    experiment_resampling()  # For task 2.2.2

    print("\n" + "="*80)
    print("Running New Sinusoidal Resampling Impact Simulation...")
    print("="*80)
    simulate_resampling_impact()  # New function for the task

    # Other experiments are commented out
    """
    print("\n" + "="*80)
    print("Running Clock Recovery Experiment...")
    print("="*80)
    experiment_clock_recovery()

    print("\n" + "="*80)
    print("Running Quantization Experiment...")
    print("="*80)
    experiment_quantization()

    print("\n" + "="*80)
    print("Running Delay I/Q Experiment...")
    print("="*80)
    experiment_delay_iq()
    """

# Run all experiments when this script is executed
if __name__ == "__main__":
    run_all_experiments()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import scipy.signal as signal

def experiment_clock_recovery():
    """
    Simulate a sinusoidal signal and apply clock recovery to adjust the sampling frequency.
    Use a slight jitter in the clock recovery process and analyze its effects.
    """
    plt.figure(figsize=(15, 18))
    gs = GridSpec(5, 1)

    # Parameters for the simulation
    duration = 0.1  # seconds
    Fs_original = 1000  # Original sampling frequency: 1000 Hz

    # Carrier frequency (sinusoidal signal)
    fc = 100  # 100 Hz carrier

    # Generate original signal
    t_original = np.arange(0, duration, 1/Fs_original)
    signal_original = np.sin(2 * np.pi * fc * t_original)

    # Add some envelope variation to better visualize timing errors
    envelope = 1 + 0.3 * np.sin(2 * np.pi * 5 * t_original)  # 5 Hz modulation
    signal_original = signal_original * envelope

    # Plot the original signal
    ax1 = plt.subplot(gs[0])
    ax1.plot(t_original, signal_original, 'b-')
    ax1.set_title(f'Original Signal (Fs = {Fs_original} Hz, fc = {fc} Hz)')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Amplitude')
    ax1.grid(True)

    # Define different clock recovery scenarios
    # 1. Ideal clock recovery (no jitter)
    Fs_recovered = 800  # Recovered sampling frequency (intentionally different)
    t_recovered_ideal, signal_recovered_ideal = clockSamplingInterp(
        signal_original, Fs_original, Fs_recovered, jitter_rms=0)

    # 2. Mild jitter in clock recovery
    jitter_mild = 50e-6  # 50 microseconds RMS jitter
    t_recovered_mild, signal_recovered_mild = clockSamplingInterp(
        signal_original, Fs_original, Fs_recovered, jitter_rms=jitter_mild)

    # 3. Severe jitter in clock recovery
    jitter_severe = 200e-6  # 200 microseconds RMS jitter
    t_recovered_severe, signal_recovered_severe = clockSamplingInterp(
        signal_original, Fs_original, Fs_recovered, jitter_rms=jitter_severe)

    # Plot the ideal clock recovery result
    ax2 = plt.subplot(gs[1])
    ax2.plot(t_original, signal_original, 'b-', alpha=0.5, label='Original')
    ax2.plot(t_recovered_ideal, signal_recovered_ideal, 'g-o', markersize=4, label='Recovered (No Jitter)')
    ax2.set_xlim(0, 0.05)  # Show only first 0.05 seconds for better visibility
    ax2.set_title(f'Ideal Clock Recovery: {Fs_original} Hz  {Fs_recovered} Hz (No Jitter)')
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Amplitude')
    ax2.legend()
    ax2.grid(True)

    # Plot the mild jitter clock recovery result
    ax3 = plt.subplot(gs[2])
    ax3.plot(t_original, signal_original, 'b-', alpha=0.5, label='Original')
    ax3.plot(t_recovered_mild, signal_recovered_mild, 'r-o', markersize=4, label=f'Recovered (Jitter RMS: {jitter_mild*1e6:.0f} s)')
    ax3.set_xlim(0, 0.05)
    ax3.set_title(f'Clock Recovery with Mild Jitter: {Fs_original} Hz  {Fs_recovered} Hz')
    ax3.set_xlabel('Time (s)')
    ax3.set_ylabel('Amplitude')
    ax3.legend()
    ax3.grid(True)

    # Plot the severe jitter clock recovery result
    ax4 = plt.subplot(gs[3])
    ax4.plot(t_original, signal_original, 'b-', alpha=0.5, label='Original')
    ax4.plot(t_recovered_severe, signal_recovered_severe, 'm-o', markersize=4, label=f'Recovered (Jitter RMS: {jitter_severe*1e6:.0f} s)')
    ax4.set_xlim(0, 0.05)
    ax4.set_title(f'Clock Recovery with Severe Jitter: {Fs_original} Hz  {Fs_recovered} Hz')
    ax4.set_xlabel('Time (s)')
    ax4.set_ylabel('Amplitude')
    ax4.legend()
    ax4.grid(True)

    # Detailed view of timing errors caused by jitter
    ax5 = plt.subplot(gs[4])
    # Select a small window to see timing errors more clearly
    window_start = 0.02
    window_length = 0.01
    window_end = window_start + window_length

    mask_original = (t_original >= window_start) & (t_original <= window_end)
    mask_ideal = (t_recovered_ideal >= window_start) & (t_recovered_ideal <= window_end)
    mask_mild = (t_recovered_mild >= window_start) & (t_recovered_mild <= window_end)
    mask_severe = (t_recovered_severe >= window_start) & (t_recovered_severe <= window_end)

    ax5.plot(t_original[mask_original], signal_original[mask_original], 'b-', label='Original Signal')
    ax5.plot(t_recovered_ideal[mask_ideal], signal_recovered_ideal[mask_ideal], 'g-o', markersize=4, label='No Jitter')
    ax5.plot(t_recovered_mild[mask_mild], signal_recovered_mild[mask_mild], 'r-o', markersize=4, label=f'Mild Jitter ({jitter_mild*1e6:.0f} s RMS)')
    ax5.plot(t_recovered_severe[mask_severe], signal_recovered_severe[mask_severe], 'm-o', markersize=4, label=f'Severe Jitter ({jitter_severe*1e6:.0f} s RMS)')

    ax5.set_xlim(window_start, window_end)
    ax5.set_title('Detailed View: Effect of Clock Jitter on Sampling Times')
    ax5.set_xlabel('Time (s)')
    ax5.set_ylabel('Amplitude')
    ax5.legend()
    ax5.grid(True)

    plt.tight_layout()
    plt.savefig('clock_recovery_jitter_simulation.png', dpi=300)
    plt.show()

    # Calculate and display clock recovery metrics
    print("\n=== Clock Recovery Simulation Results ===")
    print(f"Original signal: Fs = {Fs_original} Hz, fc = {fc} Hz")
    print(f"Recovered signal frequency: {Fs_recovered} Hz")
    print(f"Ratio of recovered to original sampling rate: {Fs_recovered/Fs_original:.3f}")

    # Calculate timing deviations caused by jitter
    ideal_timing = np.arange(len(t_recovered_ideal)) / Fs_recovered

    # Calculate timing error statistics for the mild jitter case
    timing_error_mild = t_recovered_mild - ideal_timing[:len(t_recovered_mild)]
    print(f"\nMild Jitter ({jitter_mild*1e6:.1f} s RMS) Effects:")
    print(f"  Mean timing error: {np.mean(timing_error_mild)*1e6:.2f} s")
    print(f"  Std dev of timing error: {np.std(timing_error_mild)*1e6:.2f} s")
    print(f"  Max timing error: {np.max(np.abs(timing_error_mild))*1e6:.2f} s")

    # Calculate timing error statistics for the severe jitter case
    timing_error_severe = t_recovered_severe - ideal_timing[:len(t_recovered_severe)]
    print(f"\nSevere Jitter ({jitter_severe*1e6:.1f} s RMS) Effects:")
    print(f"  Mean timing error: {np.mean(timing_error_severe)*1e6:.2f} s")
    print(f"  Std dev of timing error: {np.std(timing_error_severe)*1e6:.2f} s")
    print(f"  Max timing error: {np.max(np.abs(timing_error_severe))*1e6:.2f} s")

    # Calculate and display signal quality metrics
    # Reconstruct the signals to the original time base for comparison
    reconstructed_ideal = np.interp(t_original, t_recovered_ideal, signal_recovered_ideal)
    reconstructed_mild = np.interp(t_original, t_recovered_mild, signal_recovered_mild)
    reconstructed_severe = np.interp(t_original, t_recovered_severe, signal_recovered_severe)

    # Calculate SNR (Signal-to-Noise Ratio) - using the difference as noise
    def calculate_snr(original, recovered):
        noise = original - recovered
        signal_power = np.mean(original**2)
        noise_power = np.mean(noise**2)
        if noise_power > 0:
            snr = 10 * np.log10(signal_power / noise_power)
        else:
            snr = float('inf')
        return snr

    snr_ideal = calculate_snr(signal_original, reconstructed_ideal)
    snr_mild = calculate_snr(signal_original, reconstructed_mild)
    snr_severe = calculate_snr(signal_original, reconstructed_severe)

    print("\nSignal Quality Metrics:")
    print(f"  SNR with ideal clock recovery: {snr_ideal:.2f} dB")
    print(f"  SNR with mild jitter: {snr_mild:.2f} dB")
    print(f"  SNR with severe jitter: {snr_severe:.2f} dB")

    # Discussion of clock recovery and jitter impact
    print("\n=== Discussion: Effect of Jitter on Clock Recovery ===")
    print("1. Clock Recovery Process:")
    print("   - Clock recovery involves resampling a signal from one clock domain to another")
    print("   - It's essential for synchronizing digital systems with different clock rates")
    print("   - The process uses interpolation to estimate signal values at new sampling instants")

    print("\n2. Jitter Effects:")
    print("   - Clock jitter introduces random variations in the sampling times")
    print("   - This causes amplitude errors in the recovered signal")
    print("   - Higher frequency components are more sensitive to timing jitter")
    print(f"   - Even mild jitter ({jitter_mild*1e6:.1f} s RMS) reduced SNR by {snr_ideal-snr_mild:.2f} dB")
    print(f"   - Severe jitter ({jitter_severe*1e6:.1f} s RMS) reduced SNR by {snr_ideal-snr_severe:.2f} dB")

    print("\n3. Practical Implications:")
    print("   - In communication systems, jitter directly impacts Bit Error Rate (BER)")
    print("   - Clock recovery circuits must minimize jitter to maintain signal integrity")
    print("   - PLL (Phase-Locked Loop) or DLL (Delay-Locked Loop) designs try to minimize jitter")
    print("   - Systems with higher bandwidth requirements are more sensitive to jitter effects")
    print("   - Clock jitter specifications are critical for high-speed serial interfaces")

# Enhanced clockSamplingInterp function with more detailed implementation
def clockSamplingInterp(signal, Fs_in, Fs_out, jitter_rms=0):
    """
    Perform clock sampling interpolation on a signal

    Parameters:
    ----------
    signal : numpy array
        Input signal to be resampled
    Fs_in : float
        Input sampling frequency
    Fs_out : float
        Output sampling frequency
    jitter_rms : float
        RMS jitter to add to sampling times in seconds

    Returns:
    -------
    tuple (numpy array, numpy array)
        (sampling times, resampled signal with adjusted clock)
    """
    # Create time vector for input signal
    t_in = np.arange(len(signal)) / Fs_in

    # Calculate the number of samples in the output signal
    t_duration = t_in[-1]
    n_samples_out = int(np.ceil(t_duration * Fs_out))

    # Create ideal sampling times
    t_out_ideal = np.arange(n_samples_out) / Fs_out

    # Add jitter if requested
    if jitter_rms > 0:
        # Generate random timing jitter with specified RMS value
        jitter = np.random.normal(0, jitter_rms, n_samples_out)
        t_out = t_out_ideal + jitter

        # Ensure monotonically increasing time points
        # (real hardware would have physical constraints that enforce this)
        for i in range(1, len(t_out)):
            if t_out[i] <= t_out[i-1]:
                t_out[i] = t_out[i-1] + 1e-10  # Add tiny increment to preserve order
    else:
        t_out = t_out_ideal

    # Only keep samples within input time range
    valid_mask = t_out <= t_in[-1]
    t_out = t_out[valid_mask]

    # Interpolate signal values at new sampling times
    resampled_signal = np.interp(t_out, t_in, signal)

    return t_out, resampled_signal

# Run the clock recovery experiment when this script is executed
if __name__ == "__main__":
    experiment_clock_recovery()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import scipy.signal as signal

def delaySignal(signal_data, delay_seconds, sample_rate):
    """
    Apply a time delay to a signal

    Args:
        signal_data: Input signal array
        delay_seconds: Delay in seconds
        sample_rate: Sample rate in Hz

    Returns:
        Delayed signal
    """
    # Convert delay in seconds to samples
    delay_samples = int(delay_seconds * sample_rate)

    # Create delayed signal (zero-padded)
    delayed_signal = np.zeros_like(signal_data)

    # Shift the signal by the delay amount
    if delay_samples > 0:
        delayed_signal[delay_samples:] = signal_data[:-delay_samples]
    else:
        # For negative delays (advance), which we don't use in this example
        delayed_signal = signal_data

    return delayed_signal

def experiment_delay_iq():
    """
    Apply delay to a complex signal (real and imaginary parts) and plot both components.
    Analyze the impact of delay on I/Q components and discuss how this relates to
    synchronization errors and timing skew between I and Q channels.
    """
    plt.figure(figsize=(15, 20))
    gs = GridSpec(7, 1)

    # Parameters for the simulation
    duration = 0.1  # seconds
    Fs = 10000  # Sampling frequency: 10 kHz

    # Generate time vector
    t = np.arange(0, duration, 1/Fs)
    num_samples = len(t)

    # Create a complex modulated signal
    # Carrier frequency
    fc = 100  # 100 Hz carrier

    # Generate baseband signals (lower frequency than carrier)
    fb_I = 10  # 10 Hz baseband for I
    fb_Q = 15  # 15 Hz baseband for Q (different to show phase relationships better)

    # Create I and Q baseband components
    I_baseband = np.cos(2 * np.pi * fb_I * t)
    Q_baseband = np.sin(2 * np.pi * fb_Q * t)

    # Create complex baseband signal
    z_baseband = I_baseband + 1j * Q_baseband

    # Modulate to create bandpass signal
    z_modulated = z_baseband * np.exp(1j * 2 * np.pi * fc * t)

    # Extract I and Q components from modulated signal
    I_modulated = np.real(z_modulated)
    Q_modulated = np.imag(z_modulated)

    # Plot the original I/Q components
    ax1 = plt.subplot(gs[0])
    ax1.plot(t, I_modulated, 'b-', label='I (Real)')
    ax1.plot(t, Q_modulated, 'r-', label='Q (Imaginary)')
    ax1.set_title('Original I/Q Modulated Signal (No Delay)')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Amplitude')
    ax1.legend()
    ax1.grid(True)

    # Create zoomed view of original signal
    ax2 = plt.subplot(gs[1])
    zoom_start = 0.01
    zoom_duration = 0.02
    zoom_end = zoom_start + zoom_duration
    zoom_mask = (t >= zoom_start) & (t <= zoom_end)

    ax2.plot(t[zoom_mask], I_modulated[zoom_mask], 'b-', label='I (Real)')
    ax2.plot(t[zoom_mask], Q_modulated[zoom_mask], 'r-', label='Q (Imaginary)')
    ax2.set_title('Zoomed View of Original I/Q Signal')
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Amplitude')
    ax2.legend()
    ax2.grid(True)

    # Define several delay scenarios and apply them
    delays = [
        (1/Fs, "One Sample Delay (I delayed)"),              # One sample delay
        (0.25/fc, "Quarter Cycle Delay (I delayed)"),        # 1/4 of carrier cycle
        (0.5/fc, "Half Cycle Delay (I delayed)"),            # 1/2 of carrier cycle
        (1.0/fc, "Full Cycle Delay (I delayed)")             # Full carrier cycle
    ]

    # Process each delay scenario
    constellation_data = []

    for i, (delay, title) in enumerate(delays):
        # Apply delay to I component only (to simulate I/Q timing skew)
        I_delayed = delaySignal(I_modulated, delay, Fs)
        Q_unchanged = Q_modulated

        # Reconstruct complex signal with delayed I
        z_skewed = I_delayed + 1j * Q_unchanged

        # Save some constellation points for later plotting
        if i == 0:  # Save only for the first delay scenario to avoid clutter
            # Sample at regular intervals
            sample_indices = np.linspace(0, num_samples-1, 100).astype(int)
            constellation_data.append((I_modulated[sample_indices], Q_modulated[sample_indices], "Original"))
            constellation_data.append((I_delayed[sample_indices], Q_unchanged[sample_indices], f"Delayed ({delay*1000:.2f} ms)"))

        # Plot time domain comparison for this delay
        ax = plt.subplot(gs[i+2])
        ax.plot(t[zoom_mask], I_modulated[zoom_mask], 'b-', alpha=0.5, label='I (Original)')
        ax.plot(t[zoom_mask], I_delayed[zoom_mask], 'b--', label=f'I (Delayed by {delay*1000:.2f} ms)')
        ax.plot(t[zoom_mask], Q_unchanged[zoom_mask], 'r-', label='Q (Unchanged)')
        ax.set_title(f'I/Q Timing Skew: {title}')
        ax.set_xlabel('Time (s)')
        ax.set_ylabel('Amplitude')
        ax.legend()
        ax.grid(True)

        # Calculate phase difference between original and skewed signals
        # We need to handle complex signals differently than in the original code

        # Calculate instantaneous phase for both signals
        phase_original = np.angle(z_modulated)
        phase_skewed = np.angle(z_skewed)

        # Compute phase difference
        phase_diff = phase_skewed - phase_original
        phase_diff = np.rad2deg(np.unwrap(phase_diff))  # Convert to degrees and unwrap

        # If this is the last delay scenario, plot phase difference
        if i == len(delays) - 1:
            ax_phase = plt.subplot(gs[i+3])
            ax_phase.plot(t, phase_diff, 'g-')
            ax_phase.set_title(f'Phase Difference Due to I/Q Timing Skew ({delay*1000:.2f} ms delay)')
            ax_phase.set_xlabel('Time (s)')
            ax_phase.set_ylabel('Phase Difference (degrees)')
            ax_phase.grid(True)

            # Calculate and display average phase difference
            avg_phase_diff = np.mean(phase_diff)
            ax_phase.axhline(y=avg_phase_diff, color='r', linestyle='--', label=f'Avg: {avg_phase_diff:.2f}')
            ax_phase.legend()

    # Create a new figure for constellation diagrams
    plt.figure(figsize=(12, 6))

    # Plot the constellation diagrams
    for i, (I_data, Q_data, label) in enumerate(constellation_data):
        plt.subplot(1, 2, i+1)
        plt.scatter(I_data, Q_data, alpha=0.7, label=label)
        plt.title(f'Constellation Diagram: {label}')
        plt.xlabel('I (Real)')
        plt.ylabel('Q (Imaginary)')
        plt.grid(True)
        plt.axis('equal')  # Equal scaling to maintain shape
        plt.legend()

    plt.tight_layout()
    plt.savefig('iq_delay_constellation.png', dpi=300)

    # Show the main figure
    plt.figure(1)
    plt.tight_layout()
    plt.savefig('iq_delay_analysis.png', dpi=300)
    plt.show()

    # Calculate and display metrics related to I/Q timing skew
    print("\n=== I/Q Delay and Timing Skew Analysis ===")
    print(f"Original signal: Fs = {Fs} Hz, fc = {fc} Hz")
    print(f"I baseband frequency: {fb_I} Hz, Q baseband frequency: {fb_Q} Hz")

    print("\nTiming Skew Scenarios:")
    for delay, title in delays:
        # Calculate skew as a fraction of carrier period
        skew_fraction = delay * fc

        # Calculate phase offset in degrees
        phase_offset = skew_fraction * 360

        print(f"\n{title}:")
        print(f"  Delay: {delay*1e6:.2f} s")
        print(f"  As fraction of carrier period: {skew_fraction:.4f}")
        print(f"  Equivalent phase offset: {phase_offset:.2f}")

        # Calculate expected error types based on delay amount
        if skew_fraction < 0.1:
            impact = "Minor phase distortion, small constellation rotation"
        elif skew_fraction < 0.25:
            impact = "Moderate distortion, noticeable constellation rotation"
        elif 0.45 < skew_fraction < 0.55:
            impact = "Severe distortion, approximately 180 phase shift (sign reversal)"
        elif skew_fraction > 0.9 and skew_fraction < 1.1:
            impact = "Minimal distortion if delay is exactly one period, otherwise phase rotation"
        else:
            impact = "Significant distortion, substantial constellation rotation"

        print(f"  Expected impact: {impact}")

    # Discussion of I/Q delay effects
    print("\n=== Discussion: Impact of I/Q Timing Skew ===")
    print("1. I/Q Timing Skew Fundamentals:")
    print("   - I/Q timing skew occurs when I and Q components are not perfectly time-aligned")
    print("   - Common causes: Unequal path lengths, component variations, clock distribution issues")
    print("   - Effects vary with skew magnitude relative to signal period")

    print("\n2. Effects on Signal Constellation:")
    print("   - Small skew: Slight rotation of constellation points")
    print("   - Quarter cycle skew (90): Circular constellation becomes elliptical")
    print("   - Half cycle skew (180): Can invert signal components")
    print("   - Full cycle skew: May appear normal but causes phase ambiguity")

    print("\n3. System Performance Impact:")
    print("   - Reduced error vector magnitude (EVM) and signal-to-noise ratio")
    print("   - Higher bit error rate (BER) in digital communications")
    print("   - Degraded image rejection in receivers")
    print("   - Impaired adjacent channel rejection")

    print("\n4. Relationship to Digital Communication:")
    print("   - More significant impact on higher-order modulation schemes (e.g., 64-QAM vs QPSK)")
    print("   - Modern systems typically require I/Q timing alignment within 1-2% of symbol period")
    print("   - Skew compensation techniques include:")
    print("     * Digital adaptive filtering")
    print("     * Fractional delay filters")
    print("     * Training sequence-based calibration")

    print("\n5. Practical Considerations:")
    print("   - Clock distribution networks must minimize skew")
    print("   - Component matching critical in analog front-ends")
    print("   - Temperature variations can introduce dynamic skew")
    print("   - Test equipment must account for and measure I/Q timing errors")

# Run the I/Q delay experiment when this script is executed
if __name__ == "__main__":
    experiment_delay_iq()
```
